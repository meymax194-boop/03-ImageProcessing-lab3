```python
# Import the library
from PIL import Image, ImageEnhance
import matplotlib.pyplot as plt
```

```python
# ── Load image ────────────────────────────────────
img = Image.open('cameraman.jpg')

plt.imshow(img, cmap='gray')
plt.axis('off')
```

### 1. Scale (increase size)
Double the image dimensions using `Image.resize()` with high-quality resampling `(LANCZOS)`.

```python
# Get the size of the image
width, height = img.size

cx = 2
cy = 2
# scaling factors
```

```python
# ── 1. Increase size (scale ×2) ────────────────────

# Resize the image using PIL's built-in method
new_width = int(width * cx)
new_height = int(height * cy)

scaled_img = img.resize((new_width, new_height), Image.LANCZOS)

plt.imshow(scaled_img, cmap='gray')
plt.title('Scaled x2')
plt.axis('off')
```

```python

# Save the scaled image and print the sizes (The new image name should be "task1_1_scaled.jpg")
scaled_img.save('task1_1_scaled.jpg')

print("Original size:", img.size)
print("Scaled size:", scaled_img.size)
```

 non-uniform scale (cx=2, cy=1) → stretch horizontally only

```python
cx = 2
cy = 1
```

```python
new_width = int(width * cx)
new_height = int(height * cy)

non_uniform_img = img.resize((new_width, new_height), Image.LANCZOS)

plt.imshow(non_uniform_img, cmap='gray')
plt.title('Non-uniform scaling (horizontal stretch)')
plt.axis('off')
```

```python
non_uniform_img.save('task1_2_nonuniform.jpg')

print("Non-uniform size:", non_uniform_img.size)
```

### 2. Rotate 120°
Rotate the image by 120 degrees, expanding the canvas to fit the full rotated image.

```python
# ── 2. Rotate 120 degrees ──────────────────────────
# 2. Rotate 120 degrees

rotated_img = img.rotate(120, expand=True)

plt.imshow(rotated_img, cmap='gray')
plt.title('Rotated 120°')
plt.axis('off')
# Save the scaled image and print the sizes (The new image name should be "task1_2_rotated.jpg")
rotated_img.save('task_2_rotated.jpg')

print("Original size:", img.size)
print("Rotated size:", rotated_img.size)
```

### 3. Shear

```python
# -- c. Get the image dimensions ────────────────────
width, height = img.size
```

```python
# -- d. define the shear matrix ──────────────────────────
# Choose X-axis or Y-axis shear. The shear factor controls how much the image slants — start with 0.5 then experiment.
shx = 0.5   # shear factor on x-axis
shy = 0     # no shear on y-axis
```

```python
# -- e. Apply the shear transformation to the image ──────────────────────────
# PIL's transform() takes the inverse affine matrix:
# [ 1    shx   tx ]
# [ shy  1     ty ]
sheared_img = img.transform(
    (width, height),
    Image.AFFINE,
    (1, shx, 0,
     shy, 1, 0)
)

plt.imshow(sheared_img, cmap='gray')
plt.title('Sheared Image')
plt.axis('off')
```

```python
# -- f. Save the sheared image(The new image name should be "task1_3_sheared.jpg") 
sheared_img.save('task_3_sheared.jpg')

print("Sheared size:", sheared_img.size)
```

# Intensity Transformations
Negative · Log · Power Law (Gamma)

```python

# ── 1. Negative ──────────────────────────────────── 
# Method 1: NumPy array manipulation


import numpy as np
from PIL import ImageOps
arr = np.array(img)

negative_arr = 255 - arr
negative_img = Image.fromarray(np.uint8(negative_arr))

plt.imshow(negative_img, cmap='gray')
plt.title('Negative Image')
plt.axis('off')
```

```python
# Method 2: PIL's ImageOps
negative_img_pil = ImageOps.invert(img)

plt.imshow(negative_img_pil, cmap='gray')
plt.title('Negative Image using ImageOps')
plt.axis('off')
```

```python
# ── 2. Log transformation ──────────────────────────

# s = c · log(1 + r) --> We need to find c.
# We want the maximum output value (s) to be 255 when the maximum input value (r) is 255:
# 255 = c · log(1 + 255)
# c = 255 / log(1 + 255)
arr = np.array(img, dtype=float)

c = 255 / np.log(1 + 255)

log_transformed_arr = c * np.log(1 + arr)
log_transformed_img = Image.fromarray(np.uint8(log_transformed_arr))

plt.imshow(log_transformed_img, cmap='gray')
plt.title('Log Transformation')
plt.axis('off')


# Apply the log transformation to each pixel
```

```python

# ── 3. Power-law / Gamma correction ───────────────
arr = np.array(img, dtype=float)

gamma = 0.5

gamma_corrected_arr = 255 * ((arr / 255) ** gamma)
gamma_corrected_img = Image.fromarray(np.uint8(gamma_corrected_arr))

plt.imshow(gamma_corrected_img, cmap='gray')
plt.title('Gamma Correction')
plt.axis('off')
```

```python
gamma_corrected_img.save('task2_3_gamma.jpg')

print("Gamma corrected size:", gamma_corrected_img.size)
```
